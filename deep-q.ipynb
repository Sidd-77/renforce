```python
import gym
import tf
from gym.spaces import Discrete, Dict
from gym.utils.play import play
import pygame
import random
```

```python
SCREEN_WIDTH = 720
SCREEN_HEIGHT = 720
X_POSITIONS = [120, 240, 360, 480, 600]
```

```python
def display_score(score, screen):
   font=pygame.font.Font(None,30)
   scoretext=font.render("Score:"+str(score), 1,(255,255,255))
   screen.blit(scoretext, (620, 20))
```

```python
class stones:
    def __init__(self) -> None:
        self.y = 0
        self.x = random.choice(X_POSITIONS)

    def fall(self):
        self.y += 5
```

```python
class player:
    def __init__(self) -> None:
        self.x = 360
        self.y = 600

    def moveLeft(self):
        if(self.x > 120):
            self.x -= 120


    def moveRight(self):
        if(self.x < 600):
            self.x += 120

    def collide(self, stone:stones):
        dist = (stone.x - self.x)**2 + (stone.y - self.y)**2
        ideal_dist = 40**2
        if(dist <= ideal_dist):
            return True
```

```python
from typing import List


class MyEnv(gym.Env):

    def __init__(self) -> None:
        super().__init__()

        self.state = 3

        self.observation_space = Dict({
            "agent": Discrete(5, start=1),
            "player": Discrete(5, start=1),
        })

        self.action_space = Discrete(3, start=-1)

        self._action_taken = {
            -1: "moveleft",
            0: "stay",
            1: "moveright"
        }

        self.stones_list = []
        self.p1 = player()
        self.player_score = 0;

    def step(self, action):

        self.reward = 0
        self.done = False

        if action == 0:
            pass
        elif action == 1:
            self.p1.moveRight()
            self.state -= 1
            if(self.state == 0):
                self.state = 5
        elif action == -1:
            self.p1.moveLeft()
            self.state += 1
            if(self.state == 6):
                self.state = 1

        for stone in self.stones_list:
            stone.fall()
            
            if (stone.y > 720):
                self.player_score += 1
                self.reward += 1
                self.stones_list.remove(stone)

            if self.p1.collide(stone):
                print("Collided")
                print("Score: ", self.player_score)
                self.running = False
                self.done = True
                self.reward -= 100

        return self.state, self.reward, self.done, {}

    def reset(self):
        self.stones_list = []
        self.p1 = player()
        self.player_score = 0
        self.state = 3
        print(self.state)
        return {'agent':5 ,'player': self.state}, {}
    
    def render(self) :
        pygame.init()
        self.screen = pygame.display.set_mode((720, 720))
        self.clock = pygame.time.Clock()
        self.running = True

        self.stones_list = []

        self.CREATE_STONE = pygame.USEREVENT + 1
        pygame.time.set_timer(self.CREATE_STONE, 1000)

        self.ALLOW_INPUT = pygame.USEREVENT + 2
        self.accepting_input = True

        for stone in self.stones_list:
            stone.obj = pygame.draw.circle(self.screen, "red", [stone.x, stone.y], 40)
        for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    running = False

                if event.type == self.CREATE_STONE:
                    obj1 = stones()
                    self.stones_list.append(obj1)
                    obj2 = stones()
                    self.stones_list.append(obj2)

                if event.type == self.ALLOW_INPUT:
                    self.accepting_input = True

        self.screen.fill("black")

        self.player = pygame.draw.circle(self.screen, "green", [self.p1.x, self.p1.y], 40)
        pygame.display.flip()

        self.clock.tick(60)  # limits FPS to 60

        pygame.quit()
```

```python
env = MyEnv()
tp = env.observation_space.seed(1)
tp
```

```python
from gym.utils.env_checker import check_env

env = MyEnv()


while True:
    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
    print("Action: ", action)
    print("Obs: ", obs)
    print("Reward: ", reward)
    print("Done: ", done)
    print("Info: ", info)
    if done:
        break
    env.render()

env.close()

```
